```python
import pickle
import random
import sys
import time
import pygame

from pygame import mixer

# Initialize pygame
pygame.init()

# Set-up Screen
screen_width = 1000
screen_height = 500
screen = pygame.display.set_mode((screen_width, screen_height))
surface = pygame.Surface((screen_width, screen_height), pygame.SRCALPHA)  # for semitransparent bg when pause
pygame.display.set_caption('Arithmetic Spellscape')

# Font
title_font = pygame.font.Font("joystix monospace.otf", 50)
main_font = pygame.font.Font("joystix monospace.otf", 20)
medium_font = pygame.font.Font("joystix monospace.otf", 18)
smaller_font = pygame.font.Font("joystix monospace.otf", 15)
question_font = pygame.font.Font("joystix monospace.otf", 75)
answer_font = pygame.font.Font("joystix monospace.otf", 30)

# Colors
white_color = (255, 255, 255)
red_color = (255, 0, 0)
gold_color = (255, 205, 0)
black_color = (0, 0, 0)
purple_color = (66, 38, 169)

# Sound Effects
button_sfx = pygame.mixer.Sound("button_sfx.mp3")
mixer.music.load('main_music.wav')
mixer.music.play(-1)
correct_ans_sfx = pygame.mixer.Sound("correct_answer_sfx.mp3")
wrong_answer_sfx = pygame.mixer.Sound("wrong_answer_sfx.mp3")
winner_sfx = pygame.mixer.Sound("winner_sfx.wav")

# Loading Images
button_surface_new = pygame.image.load("buttonpixel.png")
button_surface_new = pygame.transform.scale(button_surface_new, (200, 75))

button_surface_old = pygame.image.load("buttonpixel1.png")
button_surface_old = pygame.transform.scale(button_surface_old, (200, 75))

forest_bg = pygame.image.load("forestbg.png")

username_text = pygame.image.load("buttonpixel.png")
username_text = pygame.transform.scale(username_text, (500, 75))

wizard_cat = pygame.image.load("wizard_cat_left.png")
wizard_cat = pygame.transform.scale(wizard_cat, (100, 100))

speech_bubble = pygame.image.load("speechbubble.png")
speech_bubble = pygame.transform.scale(speech_bubble, (600, 120))

other_dialogue_box = pygame.image.load("other_dialogue_box.png")
other_dialogue_box = pygame.transform.scale(other_dialogue_box, (other_dialogue_box.get_width() / 2, other_dialogue_box.get_height() / 3))

menu_surface_button = pygame.image.load("menu_button.png")
menu_surface_button = pygame.transform.scale(menu_surface_button, (menu_surface_button.get_width() * 1.5, menu_surface_button.get_height() * 1.5))

button_surface = pygame.image.load("buttonpixel.png")
button_surface = pygame.transform.scale(button_surface, (250, 75))

spellbook = pygame.image.load("spellbook.png")
spellbook = pygame.transform.scale(spellbook, (220, 220))


# Button
class Button():

    def __init__(self, image, x_pos, y_pos, text_input):
        self.image = image
        self.x_pos = x_pos
        self.y_pos = y_pos
        self.rect = self.image.get_rect(center=(self.x_pos, self.y_pos))
        self.text_input = text_input
        self.text = main_font.render(self.text_input, True, "black")
        self.text_rect = self.text.get_rect(center=(self.x_pos, self.y_pos))
        self.clear_screen = False
        self.show = True
        self.pause_state = False
        self.redraw = False

    def update(self):
        screen.blit(self.image, self.rect)
        screen.blit(self.text, self.text_rect)

    def checkForInput(self, position):
        if self.rect.collidepoint(position):
            self.clear_screen = True
            home_page = False

    def main_menu_input(self, position):
        if self.rect.collidepoint(position):
            self.clear_screen = True

    def pause(self, position):
        if self.rect.collidepoint(position):
            print("Flag set to True inside checkForInput")
            self.clear_screen = True
            # when menu button is pressed
            if not self.pause_state:
                self.pause_state = True
                print("sets true pause def")

            # when menu button reappears
            else:
                self.pause_state = False
                print("sets false pause def")

    def changeColor(self, position):
        if self.rect.collidepoint(position):
            self.text = main_font.render(self.text_input, True, "green")
        else:
            self.text = main_font.render(self.text_input, True, "black")

    def reset(self, text=''):
        self.text = text
        self.txt_surface = main_font.render(self.text, True, pygame.Color('black'))
        self.active = False
        self.entered = False
        self.show_message = False

    def resetUsername(self):
        input_box.reset()

    def handle_home_button():
        global current_state
        current_state = MAIN_MENU


# Text input box
class TextBox():

    def __init__(self, x, y, w, h, text=''):
        self.rect = pygame.Rect(x, y, w, h)
        self.color_inactive = pygame.Color('grey')
        self.color_active = pygame.Color('white')
        self.color = self.color_inactive
        self.text = text
        self.txt_surface = main_font.render(text, True, pygame.Color('black'))
        self.active = False
        self.entered = False
        self.show_message = False  # Flag to show the message

    #for username input
    def handle_event(self, event):
        if event.type == pygame.MOUSEBUTTONDOWN:
            if self.rect.collidepoint(event.pos):
                self.active = not self.active
            else:
                self.active = False
            self.color = self.color_active if self.active else self.color_inactive

        if event.type == pygame.KEYDOWN:
            if self.active:
                if event.key == pygame.K_RETURN:

                    if new_user:
                        if self.text:
                            if self.text in used_usernames:
                                self.show_message = True  # Show message on screen
                            else:
                                print(f"New username entered: {self.text}")
                                used_usernames.add(self.text)
                                self.entered = True
                                save_usernames('used_usernames.pkl', used_usernames)  # Save usernames to file
                        else:
                            print("Please enter a username.")

                    else:
                        if self.text:
                            if self.text in used_usernames:
                                self.entered = True
                                #get saved info to load progress
                            else:
                                #smthing to print sa screen to enter valid username
                                self.show_message = True
                                print("Enter valid username")
                        else:
                            print("Please enter a username.")

                elif event.key == pygame.K_BACKSPACE:
                    self.text = self.text[:-1]
                else:
                    if len(self.text)<30:
                        self.text += event.unicode

                self.txt_surface = main_font.render(self.text, True, pygame.Color('black'))

    def update(self):
        # Adjust width to accommodate longer usernames
        width = max(400, self.txt_surface.get_width() + 10)
        self.rect.w = width

    def draw(self, screen):
        # Center horizontally
        self.rect.x = screen_width / 2 - self.rect.w / 2
        # Center vertically
        self.rect.y = screen_height / 2 - self.rect.h / 2

        # Fill the rectangle
        pygame.draw.rect(screen, pygame.Color('white'), self.rect)
        # Draw the border
        pygame.draw.rect(screen, pygame.Color('black'), self.rect, 2)
        # Draw the text surface
        screen.blit(self.txt_surface, (self.rect.x + 5, self.rect.y + 5))

        # Display message if required
        if self.show_message and new_user:
            # Position of speech bubble at the bottom left of the screen
            bubble_x = 35
            bubble_y = 340
            screen.blit(speech_bubble, (bubble_x, bubble_y))

            # Position of text inside speech bubble
            text_x = bubble_x + 55  # Adjust horizontal offset
            text_y = bubble_y + 35  # Adjust vertical offset

            # Render text surface
            message_surface = smaller_font.render(" Sorry, this username is already taken.", True, red_color)
            text_rect = message_surface.get_rect(topleft=(text_x, text_y))

            # Show the cat
            cat_x = 10
            cat_y = 400
            screen.blit(wizard_cat, (cat_x, cat_y))

            # Ensure text fits within the speech bubble
            if text_rect.bottom > bubble_y + speech_bubble.get_height() - 10:
                text_y -= text_rect.bottom - (bubble_y + speech_bubble.get_height() - 10)
            screen.blit(message_surface, (text_x, text_y))

        # for the old username if it does not exist =======================
        elif self.show_message and old_user:
            # Position of speech bubble at the bottom left of the screen
            bubble_x = 35
            bubble_y = 340
            screen.blit(speech_bubble, (bubble_x, bubble_y))

            # Position of text inside speech bubble
            text_x = bubble_x + 55  # Adjust horizontal offset
            text_y = bubble_y + 35  # Adjust vertical offset

            # Render text surface
            message_surface = smaller_font.render(" Sorry, this username does not exist.", True, red_color)
            text_rect = message_surface.get_rect(topleft=(text_x, text_y))

            # Show the cat
            cat_x = 10
            cat_y = 400
            screen.blit(wizard_cat, (cat_x, cat_y))

            # Ensure text fits within the speech bubble
            if text_rect.bottom > bubble_y + speech_bubble.get_height() - 10:
                text_y -= text_rect.bottom - (bubble_y + speech_bubble.get_height() - 10)
            screen.blit(message_surface, (text_x, text_y))

    def reset(self):
        self.text = ''
        self.txt_surface = main_font.render(self.text, True, pygame.Color('black'))
        self.active = False
        self.entered = False
        self.show_message = False

# buttons & others :,)
button_new = Button(button_surface_new, screen_width / 2, 2 * screen_height / 3, "New Player")
button_old = Button(button_surface_old, screen_width / 2, 5 * screen_height / 6, "Old Player")
menu_button = Button(menu_surface_button, 960, 35, "")
resume_button = Button(button_surface, screen_width / 2, 1 * screen_height / 3, "Resume")
home_button = Button(button_surface, screen_width / 2, 1 * screen_height / 2, "Home")
quit_button = Button(button_surface, screen_width / 2, 2 * screen_height / 3, "Quit")
input_box = TextBox(screen_width / 2 - 200, screen_height / 2, 400, 50)
title_text = title_font.render("Arithmetic Spellscape", True, gold_color)
title_rect = title_text.get_rect(center=(screen_width // 2, 2 * screen_height // 5))

# Save used_usernames to a file
def save_usernames(filename, usernames):
    try:
        with open(filename, 'wb') as f:
            pickle.dump(usernames, f)
        print(f"Usernames successfully saved to {filename}")
    except Exception as e:
        print(f"Error saving usernames to {filename}: {e}")


# Load used_usernames from a file
def load_usernames(filename):
    try:
        with open(filename, 'rb') as f:
            return pickle.load(f)
    except FileNotFoundError:
        print(f"File {filename} not found. Returning an empty set.")
        return set()
    except Exception as e:
        print(f"Error loading usernames from {filename}: {e}")
        return set()


# Save progress function
def save_progress(username, level):
    progress_data = {'username': username, 'level': level}
    try:
        with open(f'{username}_progress.pkl', 'wb') as f:
            pickle.dump(progress_data, f)
        print(f"Progress for {username} successfully saved to {username}_progress.pkl")
    except Exception as e:
        print(f"Error saving progress for {username}: {e}")


# Loading the progress
def load_progress(username):
    try:
        with open(f'{username}_progress.pkl', 'rb') as f:
            progress_data = pickle.load(f)
            return progress_data.get('level', 1)
    except FileNotFoundError:
        print(f"No progress file found for {username}. Starting at level 1.")
        return 1
    except Exception as e:
        print(f"Error loading progress for {username}: {e}")
        return 1


def handle_home_button():
    global current_state
    current_state = MAIN_MENU


def draw_main_menu():
    pygame.draw.rect(surface, (150, 150, 150, 20), [0, 0, screen_width, screen_height])
    screen.blit(surface, (0, 0))
    resume_button.update()
    home_button.update()
    quit_button.update()


# Default states
run = True
show_username_input = False
old_user = False
new_user = False
pause = False
show_welcome_message = False
home_page = True
current_state = home_page
username = ""
reset_game_state = False

# For typewriter effect for storyline
timer = pygame.time.Clock()
speed = 3
storyline_snip = medium_font.render("", True, white_color)
storyline_counter = 0
storyline_done = "no prompt yet"
storyline1_snip = medium_font.render("", True, white_color)
storyline1_counter = 0
storyline2_snip = medium_font.render("", True, white_color)
storyline2_counter = 0
storyline3_snip = medium_font.render("", True, white_color)
storyline3_counter = 0
storyline4_snip = medium_font.render("", True, white_color)
storyline4_counter = 0
storyline4_done = False

# Track used usernames
used_usernames = load_usernames('used_usernames.pkl')
saved_progress = load_progress(f'{username}_progress.pkl')


# Function to reset game state
def reset_game():
    global addition_score, addition_streak, addition_lives
    global subtraction_score, subtraction_streak, subtraction_lives
    global multiplication_score, multiplication_streak, multiplication_lives
    global division_score, division_streak, division_lives
    global mdas_score, mdas_streak, mdas_lives

    activate_streak = False

    # Reset addition metrics
    addition_score = 0
    addition_streak = 0
    addition_lives = 9
    addition_lives_text = answer_font.render(str(addition_lives), True, black_color)
    screen.blit(addition_lives_text, (80, 30))

    # Reset subtraction metrics
    subtraction_score = 0
    subtraction_streak = 0
    subtraction_lives = 9
    subtraction_lives_text = answer_font.render(str(subtraction_lives), True, black_color)

    # Reset multiplication metrics
    multiplication_score = 0
    multiplication_streak = 0
    multiplication_lives = 9
    multiplication_lives_text = answer_font.render(str(multiplication_lives), True, black_color)

    # Reset division metrics
    division_score = 0
    division_streak = 0
    division_lives = 9
    division_lives_text = answer_font.render(str(division_lives), True, black_color)

    # Reset MDAS metrics
    mdas_score = 0
    mdas_streak = 0
    mdas_lives = 9  # Reset lives or any relevant initial value
    mdas_lives_text = answer_font.render(str(mdas_lives), True, black_color)

    print('Game Reset')
    print('Addition Lives: ', addition_lives)

    pygame.display.flip()

########################### LEVELS ###########################
# Levels Background Image
addition_bg = pygame.image.load('addition_bg.jpg')
addition_bg = pygame.transform.scale(addition_bg, (screen_width, screen_height))

subtraction_bg = pygame.image.load('subtraction_bg.jpg')
subtraction_bg = pygame.transform.scale(subtraction_bg, (screen_width, screen_height))

multiplication_bg = pygame.image.load('multiplication_bg.png')
multiplication_bg = pygame.transform.scale(multiplication_bg, (screen_width, screen_height))

division_bg = pygame.image.load('division_bg.png')
division_bg = pygame.transform.scale(division_bg, (screen_width, screen_height))

mdas_bg = pygame.image.load('mdas_bg.jpg')
mdas_bg = pygame.transform.scale(mdas_bg, (screen_width, screen_height))

# Levels Text
addition_text = title_font.render('Level 1 : Addition', False, black_color)
addition_rect = addition_text.get_rect(center=(500, 0))
pan_speed = 1  # Pixels per frame
center_y = 150  # Center at 150

subtraction_text = title_font.render('Level 2 : Subtraction', False, black_color)
subtraction_rect = subtraction_text.get_rect(center=(500, 0))
pan_speed = 1  # Pixels per frame
center_y = 150  # Center at 150

multiplication_text = title_font.render('Level 3 : Multiplication', False, black_color)
multiplication_rect = multiplication_text.get_rect(center=(500, 0))
pan_speed = 1  # Pixels per frame
center_y = 150  # Center at 150

division_text = title_font.render('Level 4 : division', False, black_color)
division_rect = division_text.get_rect(center=(500, 0))
pan_speed = 1  # Pixels per frame
center_y = 150  # Center at 150

mdas_text = title_font.render('Boss Level : MDAS', False, white_color)
mdas_rect = mdas_text.get_rect(center=(500, 0))
pan_speed = 1  # Pixels per frame
center_y = 150  # Center at 150

# Enter [x] Levels
addition_enter_text = smaller_font.render('Press [1] to start', False, black_color)
addition_enter_rect = addition_enter_text.get_rect(center=(500, 250))
blink_timer = 0
blink_rate = 500  # Blink every 500 milliseconds
is_blinking = False
addition_enter_text_alpha = 0  # Initial alpha value for visibility

subtraction_enter_text = smaller_font.render('Press [2] to start', False, black_color)
subtraction_enter_rect = subtraction_enter_text.get_rect(center=(500, 250))
subtraction_enter_text_alpha = 0  # Initial alpha value for visibility

multiplication_enter_text = smaller_font.render('Press [3] to start', False, black_color)
multiplication_enter_rect = multiplication_enter_text.get_rect(center=(500, 250))
multiplication_enter_text_alpha = 0  # Initial alpha value for visibility

division_enter_text = smaller_font.render('Press [4] to start', False, black_color)
division_enter_rect = division_enter_text.get_rect(center=(500, 250))
division_enter_text_alpha = 0  # Initial alpha value for visibility

mdas_enter_text = smaller_font.render('Press [5] to start', False, white_color)
mdas_enter_rect = mdas_enter_text.get_rect(center=(500, 250))
mdas_enter_text_alpha = 0  # Initial alpha value for visibility

# Wizard Cat Image and Settings
wizard_cat = pygame.image.load("wizard_cat_left.png")
wizard_cat = pygame.transform.scale(wizard_cat, (100, 100))
wizard_cat_x_pos = 30
wizard_target_x_pos = 1100  # Target position for the cat
wizard_moving = False  # Is the cat currently moving
wizard_move_start_time = 0
wizard_move_duration = 2000  # Duration for cat movement

# Heart Image
heart_lives = pygame.image.load("heart pixel.png")
heart_lives = pygame.transform.scale(heart_lives, (60, 60))

# Mini Boss Mouse Image and Settings
mini_boss_mouse_addition = pygame.image.load("mini_boss_mouse_addition.png")
mini_boss_mouse_addition = pygame.transform.scale(mini_boss_mouse_addition, (350, 300))
mini_boss_mouse_addition_x_pos = 650
mini_boss_mouse_addition_y_pos = 160  # Initial vertical position
mini_boss_direction = 1  # 1 for down, -1 for up

mini_boss_mouse_subtraction = pygame.image.load("mini_boss_mouse_subtraction.png")
mini_boss_mouse_subtraction = pygame.transform.scale(mini_boss_mouse_subtraction, (350, 300))
mini_boss_mouse_subtraction_x_pos = 650
mini_boss_mouse_subtraction_y_pos = 160  # Initial vertical position
mini_boss_direction = 1  # 1 for down, -1 for up

mini_boss_mouse_multiplication = pygame.image.load("mini_boss_mouse_multiplication.png")
mini_boss_mouse_multiplication = pygame.transform.scale(mini_boss_mouse_multiplication, (350, 300))
mini_boss_mouse_multiplication_x_pos = 650
mini_boss_mouse_multiplication_y_pos = 160  # Initial vertical position
mini_boss_direction = 1  # 1 for down, -1 for up

mini_boss_mouse_division = pygame.image.load("mini_boss_mouse_division.png")
mini_boss_mouse_division = pygame.transform.scale(mini_boss_mouse_division, (350, 300))
mini_boss_mouse_division_x_pos = 650
mini_boss_mouse_division_y_pos = 160  # Initial vertical position
mini_boss_direction = 1  # 1 for down, -1 for up

mini_boss_mouse_mdas = pygame.image.load("final_boss_mouse.png")
mini_boss_mouse_mdas = pygame.transform.scale(mini_boss_mouse_mdas, (350, 300))
mini_boss_mouse_mdas_x_pos = 650
mini_boss_mouse_mdas_y_pos = 160  # Initial vertical position
mini_boss_mouse_mdas_direction = 1  # 1 for down, -1 for up

# Wizard Bubble
wizard_bubble = pygame.image.load("wizard_bubble_left.png")
wizard_bubble = pygame.transform.scale(wizard_bubble, (250, 250))

# Game Elements

correct_answer_message = title_font.render('CORRECT ANSWER!', False, black_color)
show_correct_answer = False  # Flag to control display of the correct answer message
correct_answer_duration = 2000  # Duration to show the correct answer message

input_box_visible = True
input_box_hide_start_time = 0
input_box_hide_duration = wizard_move_duration

answer_input_box = pygame.Rect(400, 200, 200, 75)
user_input = ''

activate_streak = False
streak_message_shown = False

correct_answers_needed = 5

# Level 1 Elements
addition_text_finished_panning = False

mouse_addition_visible = True  # Control mouse visibility
mouse_addition_hide_start_time = 0
mouse_addition_hide_duration = wizard_move_duration

addition_question_visible = True
addition_hide_start_time = 0
addition_hide_duration = wizard_move_duration

addition_score = 0

addition_question = None
addition_answer = None

addition_streak = 0

addition_lives = 9
addition_lives_text = answer_font.render(str(addition_lives), True, black_color) # Text

# Level 2 Elements
subtraction_text_finished_panning = False

mouse_subtraction_visible = True  # Control mouse visibility
mouse_subtraction_hide_start_time = 0
mouse_subtraction_hide_duration = wizard_move_duration

subtraction_question_visible = True
subtraction_hide_start_time = 0
subtraction_hide_duration = wizard_move_duration

subtraction_score = 0

subtraction_question = None
subtraction_answer = None

subtraction_streak = 0

subtraction_lives = 9
subtraction_lives_text = answer_font.render(str(subtraction_lives), True, black_color)  # Text

# Level 3 Elements
multiplication_text_finished_panning = False

mouse_multiplication_visible = True  # Control mouse visibility
mouse_multiplication_hide_start_time = 0
mouse_multiplication_hide_duration = wizard_move_duration

multiplication_question_visible = True
multiplication_hide_start_time = 0
multiplication_hide_duration = wizard_move_duration

multiplication_score = 0

multiplication_question = None
multiplication_answer = None

multiplication_streak = 0

multiplication_lives = 9
multiplication_lives_text = answer_font.render(str(multiplication_lives), True, black_color)  # Text

# Level 4 Elements
division_text_finished_panning = False

mouse_division_visible = True  # Control mouse visibility
mouse_division_hide_start_time = 0
mouse_division_hide_duration = wizard_move_duration

division_question_visible = True
division_hide_start_time = 0
division_hide_duration = wizard_move_duration

division_score = 0

division_question = None
division_answer = None

division_streak = 0

division_lives = 9
division_lives_text = answer_font.render(str(division_lives), True, black_color)  # Text

# Boss Level Elements
mdas_text_finished_panning = False

mouse_mdas_visible = True  # Control mouse visibility
mouse_mdas_hide_start_time = 0
mouse_mdas_hide_duration = wizard_move_duration

mdas_question_visible = True
mdas_hide_start_time = 0
mdas_hide_duration = wizard_move_duration

mdas_score = 0

mdas_question = None
mdas_answer = None

mdas_streak = 0

mdas_lives = 9
mdas_lives_text = answer_font.render(str(mdas_lives), True, black_color)  # Text

# Shrink
minimum_size = 100  # Minimum size for scaling
shrink_amount = 1  # Pixels to reduce each frame

addition_current_width = mini_boss_mouse_addition.get_width()
addition_current_height = mini_boss_mouse_addition.get_height()

subtraction_current_width = mini_boss_mouse_subtraction.get_width()
subtraction_current_height = mini_boss_mouse_subtraction.get_height()

multiplication_current_width = mini_boss_mouse_multiplication.get_width()
multiplication_current_height = mini_boss_mouse_multiplication.get_height()

division_current_width = mini_boss_mouse_division.get_width()
division_current_height = mini_boss_mouse_division.get_height()

mdas_current_width = mini_boss_mouse_mdas.get_width()
mdas_current_height = mini_boss_mouse_mdas.get_height()

shrink_start_time = time.time()  # Record the start time
shrink_delay = 5  # Delay in seconds
shrink_started = False


# Levels Function
def generate_addition_question():
    num1 = random.randint(1, 99)
    num2 = random.randint(1, 99)
    question = str(num1) + " + " + str(num2) + "?"
    answer = num1 + num2
    return question, answer


def generate_subtraction_question():
    num1 = random.randint(1, 99)
    num2 = random.randint(1, num1)  # Ensure no negative results
    question = str(num1) + " - " + str(num2) + "?"
    answer = num1 - num2
    return question, answer


def generate_multiplication_question():
    num1 = random.randint(1, 15)  # Typically for simpler multiplication
    num2 = random.randint(1, 15)  # Adjust range as needed
    question = str(num1) + " × " + str(num2) + "?"
    answer = num1 * num2
    return question, answer


def generate_division_question():
    num2 = random.randint(1, 15)  # Divisor can't be zero
    answer = random.randint(1, 10)  # Random answer
    num1 = num2 * answer  # Ensure num1 is a multiple of num2
    question = f"{num1} ÷ {num2}?"  # Added space around the division symbol for clarity
    return question, answer  # Return the generated question and the answer


def generate_mdas_question():
    operation1 = random.choice(['+', '-', '×', '÷'])
    operation2 = random.choice(['+', '-', '×', '÷'])

    # Generate first pair of numbers based on the first operation
    if operation1 == '+':
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)

    elif operation1 == '-':
        num2 = random.randint(1, 10)
        num1 = random.randint(num2, 15)  # Ensure num1 is greater than num2

    elif operation1 == '×':
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)

    elif operation1 == '÷':
        num2 = random.randint(1, 10)  # Divisor can't be zero
        num1 = num2 * random.randint(1, 10)  # Ensure num1 is a multiple of num2

    # Generate third number based on the second operation
    if operation2 == '+':
        num3 = random.randint(1, 10)

    elif operation2 == '-':
        num3 = random.randint(1, 10)
        num3 = random.randint(1, num1 + 10)  # Ensure num3 is within range

    elif operation2 == '×':
        num3 = random.randint(1, 10)

    elif operation2 == '÷':
        if operation1 == '÷':
            num3 = random.randint(1, 5)
            num2 = random.randint(1, 5)
            num1 = num2 * num3 * random.randint(1, 10)  # Ensure num1 is a multiple of num2 & num 3

        else:
            num3 = random.randint(1, 5)  # Choose a reasonable range for num3
            num2 = num3 * random.randint(1, 4)  # Ensure num2 is a multiple of num3

    # Create the question string
    question = f"{num1} {operation1} {num2} {operation2} {num3}"

    # Evaluate the expression safely
    try:
        answer = eval(question.replace('÷', '/').replace('×', '*'))  # Replace for eval compatibility

        # Ensure the result is non-negative
        if answer < 0:
            return generate_mdas_question()  # Regenerate if the result is negative
    except ZeroDivisionError:
        return generate_mdas_question()  # Regenerate on division by zero

    return question, answer


# Load Levels Function
def load_level(level):
    global addition_score, subtraction_score, multiplication_score, lives, user_input
    user_input = ''
    if level == 1:
        return generate_addition_question()
    elif level == 2:
        return generate_subtraction_question()
    elif level == 3:
        return generate_multiplication_question()
    elif level == 4:
        return generate_division_question()
    elif level == 5:
        return generate_mdas_question()


# Level Completed Functions
def level_1_completed_screen():
    screen.blit(addition_bg, (0, 0))
    screen.blit(wizard_cat, (wizard_cat_x_pos, 320))
    final_score_surface = title_font.render('LEVEL 1 COMPLETED!', True, black_color)
    screen.blit(final_score_surface, ((screen.get_width() - final_score_surface.get_width()) // 2 + 10, 200))
    pygame.display.flip()
    pygame.time.delay(2000)  # Delay before resetting or transitioning to the next level


def level_2_completed_screen():
    screen.blit(subtraction_bg, (0, 0))
    screen.blit(wizard_cat, (wizard_cat_x_pos, 370))
    final_score_surface = title_font.render('LEVEL 2 COMPLETED!', True, black_color)
    screen.blit(final_score_surface, ((screen.get_width() - final_score_surface.get_width()) // 2 + 10, 200))
    pygame.display.flip()
    pygame.time.delay(2000)  # Delay before resetting or transitioning to the next level


def level_3_completed_screen():
    screen.blit(multiplication_bg, (0, 0))
    screen.blit(wizard_cat, (wizard_cat_x_pos, 390))
    final_score_surface = title_font.render('LEVEL 3 COMPLETED!', True, black_color)
    screen.blit(final_score_surface, ((screen.get_width() - final_score_surface.get_width()) // 2 + 10, 200))
    pygame.display.flip()
    pygame.time.delay(2000)  # Delay before resetting or transitioning to the next level


def level_4_completed_screen():
    screen.blit(division_bg, (0, 0))
    screen.blit(wizard_cat, (wizard_cat_x_pos, 390))
    final_score_surface = title_font.render('LEVEL 4 COMPLETED!', True, black_color)
    screen.blit(final_score_surface, ((screen.get_width() - final_score_surface.get_width()) // 2 + 10, 200))
    pygame.display.flip()
    pygame.time.delay(2000)  # Delay before resetting or transitioning to the next level


def level_5_completed_screen():
    screen.blit(mdas_bg, (0, 0))
    screen.blit(wizard_cat, (wizard_cat_x_pos, 370))
    final_score_surface = title_font.render('BOSS DEFEATED!', True, white_color)
    win_score_surface = title_font.render('YOU WIN!', True, black_color)
    screen.blit(final_score_surface, ((screen.get_width() - final_score_surface.get_width()) // 2 + 10, 200))
    screen.blit(win_score_surface, ((screen.get_width() - win_score_surface.get_width()) // 2 + 10, 300))
    pygame.display.flip()
    pygame.time.delay(2000)  # Delay before resetting or transitioning to the next level


def game_accomplished_screen():
    screen.blit(forest_bg, (0, 0))
    play_again_surface = main_font.render('> Press [P] to Play Again', True, black_color)
    restart_surface = main_font.render('> Press [Q] to Quit', True, black_color)
    spellbook_retrieved1 = answer_font.render('Thank you for retrieving', True, black_color)
    spellbook_retrieved2 = answer_font.render('the Spellbook of Wisdom!', True, black_color)
    screen.blit(spellbook_retrieved1, ((screen.get_width() - spellbook_retrieved1.get_width()) // 2 + 10, (screen.get_height() - spellbook.get_height()) // 2 - 110))
    screen.blit(spellbook_retrieved2, ((screen.get_width() - spellbook_retrieved2.get_width()) // 2 + 10, (screen.get_height() - spellbook.get_height()) // 2 - 70))
    screen.blit(play_again_surface, ((screen.get_width() - play_again_surface.get_width()) // 2 + 10, 360))
    screen.blit(restart_surface, ((screen.get_width() - play_again_surface.get_width()) // 2 + 10, 400))
    screen.blit(spellbook, ((screen.get_width() - spellbook.get_width()) // 2, (screen.get_height() - spellbook.get_height()) // 2 - 20))
    pygame.display.flip()


# Game Over Function
def game_over_screen():
    screen.fill((0, 0, 0))
    game_over_surface = title_font.render('GAME OVER', True, white_color)
    restart_surface = smaller_font.render('Press [R] to Restart', True, white_color)
    screen.blit(game_over_surface, ((screen.get_width() - game_over_surface.get_width()) // 2 + 10, screen.get_height() / 3))
    screen.blit(restart_surface, ((screen.get_width() - restart_surface.get_width()) // 2 + 10, screen.get_height() / 2))
    pygame.display.flip()


# Function to reset addition level variables
def reset_addition_level():
    global addition_score, addition_lives, user_input, addition_streak, addition_lives_text
    addition_score = 0
    addition_lives = 9  # Reset lives
    user_input = ''
    addition_streak = 0
    addition_lives_text = answer_font.render(str(addition_lives), True, black_color)


def reset_subtraction_level():
    global subtraction_score, subtraction_lives, user_input, subtraction_streak, subtraction_lives_text
    subtraction_score = 0
    subtraction_lives = 9  # Reset lives
    user_input = ''
    subtraction_streak = 0
    subtraction_lives_text = answer_font.render(str(subtraction_lives), True, black_color)


def reset_multiplication_level():
    global multiplication_score, multiplication_lives, user_input, multiplication_streak, multiplication_lives_text
    multiplication_score = 0
    multiplication_lives = 9  # Reset lives
    user_input = ''
    multiplication_streak = 0
    multiplication_lives_text = answer_font.render(str(multiplication_lives), True, black_color)


def reset_division_level():
    global division_score, division_lives, user_input, division_streak, division_lives_text
    division_score = 0
    division_lives = 9  # Reset lives
    user_input = ''
    division_streak = 0
    division_lives_text = answer_font.render(str(division_lives), True, black_color)


def reset_mdas_level():
    global mdas_score, mdas_lives, user_input, mdas_streak, mdas_lives_text
    mdas_score = 0
    mdas_lives = 9  # Reset lives
    user_input = ''
    mdas_streak = 0
    mdas_lives_text = answer_font.render(str(mdas_lives), True, black_color)


# Game States
current_state = 0
MAIN_MENU = -1
LOAD_LEVEL_1 = 1
LEVEL_1 = 2
LEVEL_1_COMPLETED = 3

LOAD_LEVEL_2 = 4
LEVEL_2 = 5
LEVEL_2_COMPLETED = 6

LOAD_LEVEL_3 = 7
LEVEL_3 = 8
LEVEL_3_COMPLETED = 9

LOAD_LEVEL_4 = 10
LEVEL_4 = 11
LEVEL_4_COMPLETED = 12

LOAD_LEVEL_5 = 13
LEVEL_5 = 14
LEVEL_5_COMPLETED = 15

GAME_OVER = 16

SHRINKING_MICE = 17

GAME_ACCOMPLISHED = 18

# Main game loop
while run:
    timer.tick(60)

    # Show main menu if pause
    if menu_button.pause_state:
        mixer.music.pause()
        draw_main_menu()

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False

        if event.type == pygame.MOUSEBUTTONDOWN and not menu_button.pause_state and button_new.show and button_old.show:
            button_new.checkForInput(pygame.mouse.get_pos())
            button_old.checkForInput(pygame.mouse.get_pos())

        if event.type == pygame.MOUSEBUTTONDOWN and not menu_button.pause_state and current_state != GAME_ACCOMPLISHED and current_state != GAME_OVER :
            menu_button.pause(pygame.mouse.get_pos())

        if event.type == pygame.MOUSEBUTTONDOWN and menu_button.pause_state:
            button_sfx.play()

            resume_button.main_menu_input(pygame.mouse.get_pos())
            if resume_button.clear_screen:
                menu_button.pause_state = False
                resume_button.clear_screen = False
                resume_button.redraw = True
                mixer.music.unpause()

            home_button.main_menu_input(pygame.mouse.get_pos())
            if home_button.clear_screen:
                button_sfx.play()
                save_progress(input_box.text, current_state)
                menu_button.pause_state = False
                home_button.clear_screen = False
                button_new.show = True
                button_old.show = True
                button_new.clear_screen = False
                button_old.clear_screen = False
                home_page = True
                show_username_input = False
                new_user = False
                old_user = False
                storyline4_done = False
                home_button.resetUsername()
                handle_home_button()
                mixer.music.unpause()
                reset_game_state = True
                print("home reset and progress saved")

            quit_button.main_menu_input(pygame.mouse.get_pos())
            if quit_button.clear_screen:
                button_sfx.play()
                run = False

        if show_username_input and not menu_button.pause_state:
            input_box.handle_event(event)

        if event.type == pygame.KEYDOWN:

            if storyline4_done and current_state != LOAD_LEVEL_1 and current_state != LOAD_LEVEL_2 and current_state != LOAD_LEVEL_3 and current_state != LOAD_LEVEL_4 and current_state != LOAD_LEVEL_5 and current_state != LEVEL_1 and current_state != LEVEL_2 and current_state != LEVEL_3 and current_state != LEVEL_4 and current_state != LEVEL_5 and current_state != LEVEL_1_COMPLETED and current_state != LEVEL_2_COMPLETED and current_state != LEVEL_3_COMPLETED and current_state != LEVEL_4_COMPLETED and current_state != LEVEL_5_COMPLETED and current_state != GAME_OVER and current_state != SHRINKING_MICE and event.key == pygame.K_SPACE:
                print('LOAD LEVEL 1')
                current_state = LOAD_LEVEL_1
                activate_streak = False
                print(activate_streak)

            if current_state == LOAD_LEVEL_1 and event.key == pygame.K_1 and addition_enter_text_alpha:
                current_state = LEVEL_1
                print("Level 1!") 

            elif current_state == LOAD_LEVEL_2 and event.key == pygame.K_2 and subtraction_enter_text_alpha:
                current_state = LEVEL_2
                print("Level 2!") 

            elif current_state == LOAD_LEVEL_3 and event.key == pygame.K_3 and multiplication_enter_text_alpha:
                current_state = LEVEL_3
                print("Level 3!") 

            elif current_state == LOAD_LEVEL_4 and event.key == pygame.K_4 and division_enter_text_alpha:
                current_state = LEVEL_4
                print("Level 4!")  

            elif current_state == LOAD_LEVEL_5 and event.key == pygame.K_5 and mdas_enter_text_alpha:
                current_state = LEVEL_5
                print("Boss Level!") 

            elif current_state == SHRINKING_MICE and event.key == pygame.K_c:
                current_state = GAME_ACCOMPLISHED
                print("Level Accomplished!") 

            elif current_state == GAME_ACCOMPLISHED:
                if event.key == pygame.K_q:
                    pygame.quit()
                    print("Quit Game!") 

                elif event.key == pygame.K_p:
                    current_state = LOAD_LEVEL_1

                    addition_score = 0
                    addition_lives = 9  # Reset lives
                    user_input = ''
                    addition_streak = 0
                    addition_lives_text = answer_font.render(str(addition_lives), True, black_color)

                    subtraction_score = 0
                    subtraction_lives = 9  # Reset lives
                    user_input = ''
                    subtraction_streak = 0
                    subtraction_lives_text = answer_font.render(str(subtraction_lives), True, black_color)

                    multiplication_score = 0
                    multiplication_lives = 9  # Reset lives
                    user_input = ''
                    multiplication_streak = 0
                    multiplication_lives_text = answer_font.render(str(multiplication_lives), True, black_color)

                    division_score = 0
                    division_lives = 9  # Reset lives
                    user_input = ''
                    division_streak = 0
                    division_lives_text = answer_font.render(str(division_lives), True, black_color)

                    mdas_score = 0
                    mdas_lives = 9  # Reset lives
                    user_input = ''
                    mdas_streak = 0
                    mdas_lives_text = answer_font.render(str(mdas_lives), True, black_color)

                    activate_streak = False

                    print("Starting Level 1!") 

            elif current_state == LEVEL_1 and event.key == pygame.K_RETURN:
                if user_input.isdigit() and int(user_input) == addition_answer:
                    correct_ans_sfx.play()
                    print('Correct!')
                    addition_score += 1
                    addition_streak += 1
                    user_input = ''  # Reset input
                    addition_question = None  # Clear current question to generate a new one

                    # Show correct answer message
                    show_correct_answer = True
                    pygame.time.set_timer(pygame.USEREVENT, correct_answer_duration)  # Set timer for duration

                    # Hide the mouse ghost when the answer is correct
                    mouse_addition_visible = False
                    mouse_addition_hide_start_time = pygame.time.get_ticks()  # Start the timer

                    # Hide the question
                    addition_question_visible = False
                    addition_hide_start_time = pygame.time.get_ticks()

                    # Hide the input box
                    input_box_visible = False
                    input_box_hide_start_time = pygame.time.get_ticks()

                    # Start cat movement
                    wizard_moving = True
                    wizard_move_start_time = pygame.time.get_ticks()
                    print("Answer:", addition_answer, "Streak:", addition_streak, "Score:", addition_score)

                    # Check if the user has answered enough questions correctly
                    if addition_score >= correct_answers_needed:
                        current_state = LEVEL_1_COMPLETED
                        winner_sfx.play()

                    if addition_streak >= 3:
                        print('Addition Streak!')
                        activate_streak = True
                        streak_message_shown = True  # Set the flag to True after showing the message

                if user_input == '':
                    print('Enter')

                elif user_input.isdigit() and int(user_input) != addition_answer:
                    user_input = ''  # Reset input for wrong answer
                    wrong_answer_sfx.play()
                    print('Wrong Answer!')
                    if not activate_streak:
                        if addition_streak < 3:  # Lose a life only if streak < 3
                            addition_lives -= 1

                    addition_streak = 0
                    activate_streak = False
                    streak_message_shown = False
                    print("Addition Lives:", addition_lives, "Addition Streak:", addition_streak)

                    addition_lives_text = answer_font.render(str(addition_lives), True, black_color)

                    if addition_lives <= 0:
                        current_state = GAME_OVER

            elif current_state == LEVEL_1:
                if event.key == pygame.K_BACKSPACE:  # Handle backspace
                    user_input = user_input[:-1]

                else:  # Allow input only if the game has started
                    user_input += event.unicode.strip()  # Append typed character

                if event.type == pygame.USEREVENT:
                    show_correct_answer = False  # Hide the correct answer message
                    pygame.time.set_timer(pygame.USEREVENT, 0)  # Stop the timer

            elif current_state == LEVEL_2 and event.key == pygame.K_RETURN:  # Check answer when Enter is pressed
                if user_input.isdigit() and int(user_input) == subtraction_answer:
                    correct_ans_sfx.play()
                    print('Correct!')
                    subtraction_score += 1
                    subtraction_streak += 1
                    user_input = ''
                    subtraction_question = None
                    show_correct_answer = True
                    pygame.time.set_timer(pygame.USEREVENT, correct_answer_duration)

                    mouse_subtraction_visible = False
                    print('Subtraction Mouse should be hidden now')
                    mouse_subtraction_hide_start_time = pygame.time.get_ticks()

                    subtraction_question_visible = False
                    subtraction_hide_start_time = pygame.time.get_ticks()

                    input_box_visible = False
                    input_box_hide_start_time = pygame.time.get_ticks()

                    wizard_moving = True
                    wizard_move_start_time = pygame.time.get_ticks()
                    print("Answer:", subtraction_answer, "Streak:", subtraction_streak, "Score:", subtraction_score)

                    if subtraction_score >= correct_answers_needed:
                        current_state = LEVEL_2_COMPLETED
                        winner_sfx.play()

                    if subtraction_streak >= 3:
                        print('Subraction Streak!')
                        activate_streak = True
                        streak_message_shown = True

                if user_input == '':
                    print('Enter')

                elif user_input.isdigit() and int(user_input) != subtraction_answer:
                    user_input = ''
                    wrong_answer_sfx.play()
                    print('Wrong Answer!')
                    if not activate_streak:
                        if subtraction_streak < 3:  # Lose a life only if streak < 3
                            subtraction_lives -= 1
                    subtraction_streak = 0
                    activate_streak = False
                    streak_message_shown = False
                    print("Subraction Lives:", subtraction_lives, "Subtraction Streak:", subtraction_streak)

                    subtraction_lives_text = answer_font.render(str(subtraction_lives), True, black_color)

                    if subtraction_lives <= 0:
                        current_state = GAME_OVER

            elif current_state == LEVEL_2:
                if event.key == pygame.K_BACKSPACE:
                    user_input = user_input[:-1]
                else:
                    user_input += event.unicode.strip()

            elif current_state == LEVEL_3 and event.key == pygame.K_RETURN:  # Check answer when Enter is pressed
                if user_input.isdigit() and int(user_input) == multiplication_answer:
                    correct_ans_sfx.play()
                    print('Correct!')
                    multiplication_score += 1
                    multiplication_streak += 1
                    user_input = ''  # Reset input
                    multiplication_question = None  # Clear current question to generate a new one

                    # Show correct answer message
                    show_correct_answer = True
                    pygame.time.set_timer(pygame.USEREVENT, correct_answer_duration)  # Set timer for duration

                    # Hide the mouse ghost when the answer is correct
                    mouse_multiplication_visible = False
                    print('Multiplication Mouse should be hidden now')
                    mouse_multiplication_hide_start_time = pygame.time.get_ticks()  # Start the timer

                    # Hide the question
                    multiplication_question_visible = False
                    multiplication_hide_start_time = pygame.time.get_ticks()

                    # Hide the input box
                    input_box_visible = False
                    input_box_hide_start_time = pygame.time.get_ticks()

                    # Start cat movement
                    wizard_moving = True
                    wizard_move_start_time = pygame.time.get_ticks()
                    print("Answer:", multiplication_answer, "Streak:", multiplication_streak, "Score:", multiplication_score)

                    # Check if the user has answered enough questions correctly
                    if multiplication_score >= correct_answers_needed:
                        current_state = LEVEL_3_COMPLETED
                        winner_sfx.play()

                    if multiplication_streak >= 3:
                        print('Multiplication Streak!')
                        activate_streak = True
                        streak_message_shown = True  # Set the flag to True after showing the message

                if user_input == '':
                    print('Enter')

                elif user_input.isdigit() and int(user_input) != multiplication_answer:
                    user_input = ''  # Reset input for wrong answer
                    wrong_answer_sfx.play()
                    print('Wrong Answer!')
                    if not activate_streak:
                        if multiplication_streak < 3:  # Lose a life only if streak < 3
                            multiplication_lives -= 1
                    multiplication_streak = 0
                    activate_streak = False
                    streak_message_shown = False
                    print("Multiplication Lives:", multiplication_lives, "Multiplication Streak:", multiplication_streak)

                    multiplication_lives_text = answer_font.render(str(multiplication_lives), True, black_color)

                    if multiplication_lives <= 0:
                        current_state = GAME_OVER

            elif current_state == LEVEL_3:
                if event.key == pygame.K_BACKSPACE:  # Handle backspace
                    user_input = user_input[:-1]

                else:  # Allow input only if the game has started
                    user_input += event.unicode.strip()

            elif current_state == LEVEL_4 and event.key == pygame.K_RETURN:  # Check answer when Enter is pressed
                if user_input.isdigit() and int(user_input) == division_answer:
                    correct_ans_sfx.play()
                    print('Correct!')
                    division_score += 1
                    division_streak += 1
                    user_input = ''
                    division_question = None
                    show_correct_answer = True
                    pygame.time.set_timer(pygame.USEREVENT, correct_answer_duration)

                    mouse_division_visible = False
                    print('Division Mouse should be hidden now')
                    mouse_division_hide_start_time = pygame.time.get_ticks()

                    division_question_visible = False
                    division_hide_start_time = pygame.time.get_ticks()

                    input_box_visible = False
                    input_box_hide_start_time = pygame.time.get_ticks()

                    wizard_moving = True
                    wizard_move_start_time = pygame.time.get_ticks()
                    print("Answer:", division_answer, "Streak:", division_streak, "Score:", division_score)
                    if division_score >= correct_answers_needed:
                        current_state = LEVEL_4_COMPLETED
                        winner_sfx.play()

                    if division_streak >= 3:
                        print('Division Streak!')
                        activate_streak = True
                        streak_message_shown = True

                if user_input == '':
                    print('Enter')

                elif user_input.isdigit() and int(user_input) != division_answer:
                    user_input = ''
                    wrong_answer_sfx.play()
                    print('Wrong Answer!')
                    if not activate_streak:
                        if division_streak < 3:  # Lose a life only if streak < 3
                            division_lives -= 1
                    division_streak = 0
                    activate_streak = False
                    streak_message_shown = False
                    print("Division Lives:", division_lives, "Division Streak:", division_streak)

                    division_lives_text = answer_font.render(str(division_lives), True, black_color)

                    if division_lives <= 0:
                        current_state = GAME_OVER

            elif current_state == LEVEL_4:
                if event.key == pygame.K_BACKSPACE:
                    user_input = user_input[:-1]
                else:
                    user_input += event.unicode.strip()

            elif current_state == LEVEL_5 and event.key == pygame.K_RETURN:  # Check answer when Enter is pressed
                if user_input.isdigit() and int(user_input) == mdas_answer:
                    correct_ans_sfx.play()
                    print('Correct!')
                    mdas_score += 1
                    mdas_streak += 1
                    user_input = ''
                    mdas_question = None
                    show_correct_answer = True
                    pygame.time.set_timer(pygame.USEREVENT, correct_answer_duration)

                    mouse_mdas_visible = False
                    print('MDAS Mouse should be hidden now')
                    mouse_mdas_hide_start_time = pygame.time.get_ticks()

                    mdas_question_visible = False
                    mdas_hide_start_time = pygame.time.get_ticks()

                    input_box_visible = False
                    input_box_hide_start_time = pygame.time.get_ticks()

                    wizard_moving = True
                    wizard_move_start_time = pygame.time.get_ticks()
                    print("Answer:", mdas_answer, "Streak:", mdas_streak, "Score:", mdas_score)
                    if mdas_score >= correct_answers_needed:
                        current_state = LEVEL_5_COMPLETED
                        winner_sfx.play()

                    if mdas_streak >= 3:
                        print('MDAS Streak!')
                        activate_streak = True
                        streak_message_shown = True

                if user_input == '':
                    print('Enter')

                elif user_input.isdigit() and int(user_input) != mdas_answer:
                    user_input = ''
                    wrong_answer_sfx.play()
                    print('Wrong Answer!')
                    if not activate_streak:
                        if mdas_streak < 3:  # Lose a life only if streak < 3
                            mdas_lives -= 1
                    mdas_streak = 0
                    activate_streak = False
                    streak_message_shown = False
                    print("MDAS Lives:", mdas_lives, "MDAS Streak:", mdas_streak)

                    mdas_lives_text = answer_font.render(str(mdas_lives), True, black_color)

                    if mdas_lives <= 0:
                        current_state = GAME_OVER

            elif current_state == LEVEL_5:
                if event.key == pygame.K_BACKSPACE:
                    user_input = user_input[:-1]
                else:
                    user_input += event.unicode.strip()

            elif current_state == GAME_OVER:
                if event.key == pygame.K_r:  # Restart on pressing 'R'
                    if addition_lives <= 0:  # If in addition level
                        current_state = LEVEL_1
                        reset_addition_level() 
                    elif subtraction_lives <= 0:
                        current_state = LEVEL_2
                        reset_subtraction_level()
                    elif multiplication_lives <= 0:
                        current_state = LEVEL_3
                        reset_multiplication_level()
                    elif division_lives <= 0:
                        current_state = LEVEL_4
                        reset_division_level()
                    elif mdas_lives <= 0:
                        current_state = LEVEL_5
                        reset_mdas_level()
                continue  # Skip the rest if in game over state

        if event.type == pygame.USEREVENT:
            show_correct_answer = False  # Hide the correct answer message
            pygame.time.set_timer(pygame.USEREVENT, 0)  # Stop the timer

    # Game Play
    if not menu_button.pause_state:
        if button_new.show and button_old.show and home_page:
            screen.blit(forest_bg, (0, 0))
            screen.blit(title_text, title_rect)
            button_new.update()
            button_old.update()
            button_new.changeColor(pygame.mouse.get_pos())
            button_old.changeColor(pygame.mouse.get_pos())

            #to reset storyline
            storyline_counter = 0
            storyline1_counter = 0
            storyline2_counter = 0
            storyline3_counter = 0
            storyline4_counter = 0
            storyline_done = "no prompt yet"

        if button_new.clear_screen:
            button_sfx.play()
            home_page = False
            if not home_page:
                screen.blit(forest_bg, (0, 0))
                button_new.show = False
                show_username_input = True
                new_user = True
                button_new.clear_screen = False 
                print("new user")

        if button_old.clear_screen:
            button_sfx.play()
            home_page = False
            if not home_page:
                screen.blit(forest_bg, (0, 0))
                button_old.show = False
                show_username_input = True
                old_user = True
                button_old.clear_screen = False
                print("old user")

        if show_username_input:
            screen.blit(forest_bg, (0, 0))
            screen.blit(username_text, (screen_width / 2 - username_text.get_width() / 2, screen_height / 2 - 150))
            username_prompt = main_font.render("Enter your username here:", True, black_color)
            screen.blit(username_prompt, (screen_width / 2 - username_prompt.get_width() / 2, screen_height / 2 - 150 + (username_text.get_height() - username_prompt.get_height()) /2))
            input_box.update()
            input_box.draw(screen)

        # For new users
        if input_box.entered and new_user and show_username_input:
            # Shows welcome message
            screen.blit(forest_bg, (0, 0))

            next_screen_text = main_font.render("Welcome, " + input_box.text + "!", True, black_color)
            screen.blit(next_screen_text, (screen_width / 2 - next_screen_text.get_width() / 2, screen_height / 2 - next_screen_text.get_height() / 2))

            show_welcome_message = True
            welcome_start_time = pygame.time.get_ticks()

            if show_welcome_message:
                while pygame.time.get_ticks() - welcome_start_time < 2000:
                    pygame.display.update()
                    for event in pygame.event.get():
                        if event.type == pygame.QUIT:
                            pygame.quit()
                            sys.exit()

            show_welcome_message = False
            show_username_input = False

            # Introduction to storyline
            screen.blit(forest_bg, (0, 0))

            screen.blit(wizard_cat, (((screen_width - wizard_cat.get_width()) / 2) + 5, 170))

            screen.blit(other_dialogue_box, (20, 190))

            wizard_cat_name = medium_font.render("Salem Whiskers the Enchantail", True, purple_color)
            screen.blit(wizard_cat_name, (30, 333))

            storyline_done = False

        # To redraw storyline bg after resume
        if new_user and not show_username_input and resume_button.redraw:
            show_welcome_message = False
            show_username_input = False

            # Introduction to storyline
            screen.blit(forest_bg, (0, 0))

            screen.blit(wizard_cat, (((screen_width - wizard_cat.get_width()) / 2) + 5, 170))

            screen.blit(other_dialogue_box, (20, 190))

            wizard_cat_name = medium_font.render("Salem Whiskers the Enchantail", True, purple_color)
            screen.blit(wizard_cat_name, (30, 333))

            resume_button.redraw = False

        storyline = input_box.text + ", my young apprentice!"
        storyline1 = "The Enchanted Forest is in chaos."
        storyline2 = "Some pesky ghost rats have stolen the Spellbook of Wisdom."
        storyline3 = "We'll need to use Arithmetic Magic to defeat them."
        storyline4 = "Hurry! We're running out of time."

        # For Intro Storyline
        if storyline_counter < speed * len(storyline) and not storyline_done and not home_page:
            storyline_counter += 1

        elif storyline_counter >= speed * len(storyline):
            storyline_done = True
            storyline1_done = False

            if storyline_done and storyline1_counter < speed * len(storyline1) and not storyline1_done:
                storyline1_counter += 1

            elif storyline1_counter >= speed * len(storyline1):
                storyline1_done = True
                storyline2_done = False

                if storyline1_done and storyline2_counter < speed * len(storyline2) and not storyline2_done:
                    storyline2_counter += 1

                elif storyline2_counter >= speed * len(storyline2):
                    storyline2_done = True
                    storyline3_done = False

                    if storyline2_done and storyline3_counter < speed * len(storyline3) and not storyline3_done:
                        storyline3_counter += 1

                    elif storyline3_counter >= speed * len(storyline3):
                        storyline3_done = True
                        storyline4_done = False

                        if storyline3_done and storyline4_counter < speed * len(storyline4) and not storyline4_done:
                            storyline4_counter += 1

                        elif storyline4_counter >= speed * len(storyline4):
                            storyline4_done = True

                        storyline4_snip = medium_font.render(storyline4[0:storyline4_counter // speed], True, black_color)
                        screen.blit(storyline4_snip, (30, 455))

                    storyline3_snip = medium_font.render(storyline3[0:storyline3_counter // speed], True, black_color)
                    screen.blit(storyline3_snip, (30, 435))

                storyline2_snip = medium_font.render(storyline2[0:storyline2_counter // speed], True, black_color)
                screen.blit(storyline2_snip, (30, 415))

            storyline1_snip = medium_font.render(storyline1[0:storyline1_counter // speed], True, black_color)
            screen.blit(storyline1_snip, (30, 395))

        storyline_snip = medium_font.render(storyline[0:storyline_counter // speed], True, black_color)
        screen.blit(storyline_snip, (30, 375))

        if new_user and storyline4_done and not home_page and not show_username_input:
            storyline_text = smaller_font.render('Enter [SPACE] to start', False, black_color)
            storyline_text_rect = storyline_text.get_rect(center=(800, 475))
            screen.blit(storyline_text, storyline_text_rect)

        # For old users
        progress_loaded = False
        if input_box.entered and old_user:
            show_username_input = False
            reset_game_state = True
            activate_streak = False
            print("Old user activate streak:", activate_streak)
            if not progress_loaded:
                addition_lives_text = answer_font.render(str(addition_lives), True, black_color)
                subtraction_lives_text = answer_font.render(str(subtraction_lives), True, black_color)
                multiplication_lives_text = answer_font.render(str(multiplication_lives), True, black_color)
                division_lives_text = answer_font.render(str(division_lives), True, black_color)
                mdas_lives_text = answer_font.render(str(mdas_lives), True, black_color)
                current_state = load_progress(input_box.text)
                if current_state == 0:
                    current_state = LOAD_LEVEL_1

                if current_state == -1:
                    current_state = LOAD_LEVEL_1
                    
                print(f"Loaded progress for {input_box.text}, starting at level {current_state}")
                progress_loaded = True
            
        if reset_game_state:
            reset_game()  # Call the reset function
            reset_game_state = False  # Reset the flag after resetting

    # Game Code
        if current_state == LOAD_LEVEL_1:
            # Move the addition text down
            if addition_rect.y < center_y:  # Only move down if above center
                addition_rect.y += pan_speed
            else:
                # Set flag to indicate text has finished panning
                addition_text_finished_panning = True

            # Update blink timer and show enter text only if panning is finished
            if addition_text_finished_panning:
                is_blinking = True
                blink_timer += timer.get_time()
                if blink_timer >= blink_rate:
                    blink_timer = 0  # Reset timer
                    # Toggle visibility by changing alpha
                    addition_enter_text_alpha = 255 if addition_enter_text_alpha == 0 else 0

            # Blit surfaces for the intro screen
            addition_lives_text = answer_font.render(str(addition_lives), True, black_color)
            screen.blit(addition_bg, (0, 0))
            screen.blit(addition_text, addition_rect)

            # Draw blinking enter text if the text has finished panning
            if addition_text_finished_panning and addition_enter_text_alpha == 255:
                addition_enter_text.set_alpha(addition_enter_text_alpha)
                screen.blit(addition_enter_text, addition_enter_rect)

        elif current_state == LEVEL_1:
            # If the game has started, draw the game elements
            screen.blit(addition_bg, (0, 0))  # Redraw the background

            # Lives
            screen.blit(heart_lives, (20, 20))
            screen.blit(addition_lives_text, (80, 30))

            # Mini Mouse Boss
            mini_boss_mouse_addition_y_pos += mini_boss_direction * 0.5  # Move up or down
            if mini_boss_mouse_addition_y_pos <= 170:  # Move up limit
                mini_boss_direction = 0.5
            elif mini_boss_mouse_addition_y_pos >= 190:  # Move down limit
                mini_boss_direction = -0.5
            # Check the elapsed time to control mouse visibility
            if not mouse_addition_visible and (
                    pygame.time.get_ticks() - mouse_addition_hide_start_time
                    >= mouse_addition_hide_duration):
                mouse_addition_visible = True
            # Draw the mini boss mouse if it's visible
            if mouse_addition_visible:
                screen.blit(mini_boss_mouse_addition, (mini_boss_mouse_addition_x_pos, mini_boss_mouse_addition_y_pos))

            # Wizard Cat
            if activate_streak:
                screen.blit(wizard_bubble, (wizard_cat_x_pos, 250))
            else:
                screen.blit(wizard_cat,
                            (wizard_cat_x_pos, 320))  # Wizard Cat Position

            if streak_message_shown:
                streak_text = answer_font.render("You're on a streak!", True, white_color)
                screen.blit(streak_text, (275, 30))

            if wizard_moving:
                current_time = pygame.time.get_ticks()  # Get current time in milliseconds
                wizard_cat_x_pos += 10  # Adjust this value for speed

                # Check if the movement duration has passed
                if current_time - wizard_move_start_time >= wizard_move_duration:
                    wizard_moving = False  # Stop moving
                    wizard_cat_x_pos = 30  # Reset to starting position
                    streak_message_shown = False

            # Question and Answer
            question_surface = question_font.render(addition_question, True, black_color)
            input_surface = title_font.render(user_input, True, black_color)

            # Addition Question
            if addition_question is None:
                addition_question, addition_answer = generate_addition_question()
            if not addition_question_visible and (pygame.time.get_ticks() - addition_hide_start_time >= addition_hide_duration):
                addition_question_visible = True
            if addition_question_visible:
                # Draw question and user input
                screen.blit(question_surface, ((screen_width - question_surface.get_width()) // 2 + 10, 75))
                screen.blit(input_surface, (answer_input_box.x + (answer_input_box.width - input_surface.get_width()) // 2, answer_input_box.y + (answer_input_box.height - input_surface.get_height()) //2))

            # Input Answer Box
            if not input_box_visible and (pygame.time.get_ticks() - input_box_hide_start_time >= input_box_hide_duration):
                input_box_visible = True
            # Render question and user input
            if input_box_visible:
                pygame.draw.rect(screen, (255, 255, 255, 20), answer_input_box, 2)

            # Draw the correct answer message if applicable
            if show_correct_answer:
                screen.blit(correct_answer_message, ((screen_width - correct_answer_message.get_width()) // 2, 150))

        elif current_state == LOAD_LEVEL_2:
            # Move the subtraction text down
            if subtraction_rect.y < center_y:  # Only move down if above center
                subtraction_rect.y += pan_speed
            else:
                # Set flag to indicate text has finished panning
                subtraction_text_finished_panning = True

            # Update blink timer and show enter text only if panning is finished
            if subtraction_text_finished_panning:
                is_blinking = True
                blink_timer += timer.get_time()
                if blink_timer >= blink_rate:
                    blink_timer = 0  # Reset timer
                    # Toggle visibility by changing alpha
                    subtraction_enter_text_alpha = 255 if subtraction_enter_text_alpha == 0 else 0

            # Blit surfaces for the intro screen
            screen.blit(subtraction_bg, (0, 0))
            screen.blit(subtraction_text, subtraction_rect)

            # Draw blinking enter text if the text has finished panning
            if subtraction_text_finished_panning and subtraction_enter_text_alpha == 255:
                subtraction_enter_text.set_alpha(subtraction_enter_text_alpha)
                screen.blit(subtraction_enter_text, subtraction_enter_rect)

        elif current_state == LEVEL_2:
            # If the game has started, draw the game elements
            screen.blit(subtraction_bg, (0, 0))  # Redraw the background

            # Lives
            screen.blit(heart_lives, (20, 20))
            screen.blit(subtraction_lives_text, (80, 30))

            # Mini Mouse Boss
            mini_boss_mouse_subtraction_y_pos += mini_boss_direction * 0.5  # Move up or down
            if mini_boss_mouse_subtraction_y_pos <= 170:  # Move up limit
                mini_boss_direction = 0.5
            elif mini_boss_mouse_subtraction_y_pos >= 190:  # Move down limit
                mini_boss_direction = -0.5
            # Check the elapsed time to control mouse visibility
            if not mouse_subtraction_visible and (pygame.time.get_ticks() - mouse_subtraction_hide_start_time >= mouse_subtraction_hide_duration):
                mouse_subtraction_visible = True

            # Draw the mini boss mouse if it's visible
            if mouse_subtraction_visible:
                screen.blit(mini_boss_mouse_subtraction, (mini_boss_mouse_subtraction_x_pos, mini_boss_mouse_subtraction_y_pos))

            # Wizard Cat
            if activate_streak:
                screen.blit(wizard_bubble, (wizard_cat_x_pos, 250))
            else:
                screen.blit(wizard_cat, (wizard_cat_x_pos, 370))  # Wizard Cat Position

            if streak_message_shown:
                streak_text = answer_font.render("You're on a streak!", True, white_color)
                screen.blit(streak_text, (275, 30))

            if wizard_moving:
                current_time = pygame.time.get_ticks()  # Get current time in milliseconds
                wizard_cat_x_pos += 10  # Adjust this value for speed

                # Check if the movement duration has passed
                if current_time - wizard_move_start_time >= wizard_move_duration:
                    wizard_moving = False  # Stop moving
                    wizard_cat_x_pos = 30  # Reset to starting position
                    streak_message_shown = False

            # Question and Answer
            question_surface = question_font.render(subtraction_question, True, black_color)
            input_surface = title_font.render(user_input, True, black_color)

            # Subtraction Question
            if subtraction_question is None:
                subtraction_question, subtraction_answer = generate_subtraction_question()
            if not subtraction_question_visible and (pygame.time.get_ticks() - subtraction_hide_start_time >= subtraction_hide_duration):
                subtraction_question_visible = True
            if subtraction_question_visible:
                # Draw question and user input
                screen.blit(question_surface, ((screen_width - question_surface.get_width()) // 2 + 10, 75))
                screen.blit(input_surface, (answer_input_box.x + (answer_input_box.width - input_surface.get_width()) // 2, answer_input_box.y + (answer_input_box.height - input_surface.get_height()) //2))

            # Input Answer Box
            if not input_box_visible and (pygame.time.get_ticks() - input_box_hide_start_time >= input_box_hide_duration):
                input_box_visible = True
            # Render question and user input
            if input_box_visible:
                pygame.draw.rect(screen, (255, 255, 255, 20), answer_input_box, 2)

            # Draw the correct answer message if applicable
            if show_correct_answer:
                screen.blit(correct_answer_message, ((screen_width - correct_answer_message.get_width()) // 2, 150))

        elif current_state == LOAD_LEVEL_3:
            # Move the multiplication text down
            if multiplication_rect.y < center_y:  # Only move down if above center
                multiplication_rect.y += pan_speed
            else:
                # Set flag to indicate text has finished panning
                multiplication_text_finished_panning = True

            # Update blink timer and show enter text only if panning is finished
            if multiplication_text_finished_panning:
                is_blinking = True
                blink_timer += timer.get_time()
                if blink_timer >= blink_rate:
                    blink_timer = 0  # Reset timer
                    # Toggle visibility by changing alpha
                    multiplication_enter_text_alpha = 255 if multiplication_enter_text_alpha == 0 else 0

            # Blit surfaces for the intro screen
            screen.blit(multiplication_bg, (0, 0))
            screen.blit(multiplication_text, multiplication_rect)

            # Draw blinking enter text if the text has finished panning
            if multiplication_text_finished_panning and multiplication_enter_text_alpha == 255:
                multiplication_enter_text.set_alpha(multiplication_enter_text_alpha)
                screen.blit(multiplication_enter_text, multiplication_enter_rect)

        elif current_state == LEVEL_3:
            # If the game has started, draw the game elements
            screen.blit(multiplication_bg, (0, 0))  # Redraw the background

            # Lives
            screen.blit(heart_lives, (20, 20))
            screen.blit(multiplication_lives_text, (80, 30))

            # Mini Mouse Boss
            mini_boss_mouse_multiplication_y_pos += mini_boss_direction * 0.5  # Move up or down
            if mini_boss_mouse_multiplication_y_pos <= 170:  # Move up limit
                mini_boss_direction = 0.5
            elif mini_boss_mouse_multiplication_y_pos >= 190:  # Move down limit
                mini_boss_direction = -0.5
            # Check the elapsed time to control mouse visibility
            if not mouse_multiplication_visible and (pygame.time.get_ticks() - mouse_multiplication_hide_start_time >= mouse_multiplication_hide_duration):
                mouse_multiplication_visible = True

            # Draw the mini boss mouse if it's visible
            if mouse_multiplication_visible:
                screen.blit(mini_boss_mouse_multiplication, (mini_boss_mouse_multiplication_x_pos, mini_boss_mouse_multiplication_y_pos))

            # Wizard Cat
            if activate_streak:
                screen.blit(wizard_bubble, (wizard_cat_x_pos, 250))
            else:
                screen.blit(wizard_cat, (wizard_cat_x_pos, 390))  # Wizard Cat Position

            if streak_message_shown:
                streak_text = answer_font.render("You're on a streak!", True, black_color)
                screen.blit(streak_text, (275, 30))

            if wizard_moving:
                current_time = pygame.time.get_ticks()  # Get current time in milliseconds
                wizard_cat_x_pos += 10  # Adjust this value for speed

                # Check if the movement duration has passed
                if current_time - wizard_move_start_time >= wizard_move_duration:
                    wizard_moving = False  # Stop moving
                    wizard_cat_x_pos = 30  # Reset to starting position
                    streak_message_shown = False

            # Question and Answer
            question_surface = question_font.render(multiplication_question, True, black_color)
            input_surface = title_font.render(user_input, True, black_color)

            # multiplication Question
            if multiplication_question is None:
                multiplication_question, multiplication_answer = generate_multiplication_question()
            if not multiplication_question_visible and (pygame.time.get_ticks() - multiplication_hide_start_time >= multiplication_hide_duration):
                multiplication_question_visible = True
            if multiplication_question_visible:
                # Draw question and user input
                screen.blit(question_surface, ((screen_width - question_surface.get_width()) // 2 + 10, 75))
                screen.blit(input_surface, (answer_input_box.x + (answer_input_box.width - input_surface.get_width()) // 2, answer_input_box.y + (answer_input_box.height - input_surface.get_height()) // 2))

            # Input Answer Box
            if not input_box_visible and (pygame.time.get_ticks() - input_box_hide_start_time >= input_box_hide_duration):
                input_box_visible = True
            # Render question and user input
            if input_box_visible:
                pygame.draw.rect(screen, (255, 255, 255, 20), answer_input_box, 2)

            # Draw the correct answer message if applicable
            if show_correct_answer:
                screen.blit(correct_answer_message, ((screen_width - correct_answer_message.get_width()) // 2, 150))

        elif current_state == LOAD_LEVEL_4:
            # Move the division text down
            if division_rect.y < center_y:  # Only move down if above center
                division_rect.y += pan_speed
            else:
                # Set flag to indicate text has finished panning
                division_text_finished_panning = True

            # Update blink timer and show enter text only if panning is finished
            if division_text_finished_panning:
                is_blinking = True
                blink_timer += timer.get_time()
                if blink_timer >= blink_rate:
                    blink_timer = 0  # Reset timer
                    # Toggle visibility by changing alpha
                    division_enter_text_alpha = 255 if division_enter_text_alpha == 0 else 0

            # Blit surfaces for the intro screen
            screen.blit(division_bg, (0, 0))
            screen.blit(division_text, division_rect)

            # Draw blinking enter text if the text has finished panning
            if division_text_finished_panning and division_enter_text_alpha == 255:
                division_enter_text.set_alpha(division_enter_text_alpha)
                screen.blit(division_enter_text, division_enter_rect)

        elif current_state == LEVEL_4:
            # If the game has started, draw the game elements
            screen.blit(division_bg, (0, 0))  # Redraw the background

            # Lives
            screen.blit(heart_lives, (20, 20))
            screen.blit(division_lives_text, (80, 30))

            # Mini Mouse Boss
            mini_boss_mouse_division_y_pos += mini_boss_direction * 0.5  # Move up or down
            if mini_boss_mouse_division_y_pos <= 170:  # Move up limit
                mini_boss_direction = 0.5
            elif mini_boss_mouse_division_y_pos >= 190:  # Move down limit
                mini_boss_direction = -0.5
            # Check the elapsed time to control mouse visibility
            if not mouse_division_visible and (
                    pygame.time.get_ticks() - mouse_division_hide_start_time
                    >= mouse_division_hide_duration):
                mouse_division_visible = True

            # Draw the mini boss mouse if it's visible
            if mouse_division_visible:
                screen.blit(mini_boss_mouse_division, (mini_boss_mouse_division_x_pos, mini_boss_mouse_division_y_pos))

            # Wizard Cat
            if activate_streak:
                screen.blit(wizard_bubble, (wizard_cat_x_pos, 250))
            else:
                screen.blit(wizard_cat, (wizard_cat_x_pos, 390))  # Wizard Cat Position

            if streak_message_shown:
                streak_text = answer_font.render("You're on a streak!", True, white_color)
                screen.blit(streak_text, (275, 30))

            if wizard_moving:
                current_time = pygame.time.get_ticks()  # Get current time in milliseconds
                wizard_cat_x_pos += 10  # Adjust this value for speed

                # Check if the movement duration has passed
                if current_time - wizard_move_start_time >= wizard_move_duration:
                    wizard_moving = False  # Stop moving
                    wizard_cat_x_pos = 30  # Reset to starting position
                    streak_message_shown = False

            # Question and Answer
            question_surface = question_font.render(division_question, True, black_color)
            input_surface = title_font.render(user_input, True, black_color)

            # division Question
            if division_question is None:
                division_question, division_answer = generate_division_question()
            if not division_question_visible and (pygame.time.get_ticks() - division_hide_start_time >= division_hide_duration):
                division_question_visible = True
            if division_question_visible:
                # Draw question and user input
                screen.blit(question_surface, ((screen_width - question_surface.get_width()) // 2 + 10, 75))
                screen.blit(input_surface, (answer_input_box.x + (answer_input_box.width - input_surface.get_width()) // 2, answer_input_box.y + (answer_input_box.height - input_surface.get_height()) //2))

            # Input Answer Box
            if not input_box_visible and (pygame.time.get_ticks() - input_box_hide_start_time >= input_box_hide_duration):
                input_box_visible = True
            # Render question and user input
            if input_box_visible:
                pygame.draw.rect(screen, (255, 255, 255, 20), answer_input_box, 2)

            # Draw the correct answer message if applicable
            if show_correct_answer:
                screen.blit(correct_answer_message, ((screen_width - correct_answer_message.get_width()) // 2, 150))

        elif current_state == LOAD_LEVEL_5:
            # Move the mdas text down
            if mdas_rect.y < center_y:  # Only move down if above center
                mdas_rect.y += pan_speed
            else:
                # Set flag to indicate text has finished panning
                mdas_text_finished_panning = True

            # Update blink timer and show enter text only if panning is finished
            if mdas_text_finished_panning:
                is_blinking = True
                blink_timer += timer.get_time()
                if blink_timer >= blink_rate:
                    blink_timer = 0  # Reset timer
                    # Toggle visibility by changing alpha
                    mdas_enter_text_alpha = 255 if mdas_enter_text_alpha == 0 else 0

            # Blit surfaces for the intro screen
            screen.blit(mdas_bg, (0, 0))
            screen.blit(mdas_text, mdas_rect)

            # Draw blinking enter text if the text has finished panning
            if mdas_text_finished_panning and mdas_enter_text_alpha == 255:
                mdas_enter_text.set_alpha(mdas_enter_text_alpha)
                screen.blit(mdas_enter_text, mdas_enter_rect)

        elif current_state == LEVEL_5:
            # If the game has started, draw the game elements
            screen.blit(mdas_bg, (0, 0))  # Redraw the background

            # Lives
            screen.blit(heart_lives, (20, 20))
            screen.blit(mdas_lives_text, (80, 30))

            # Mini Mouse Boss
            mini_boss_mouse_mdas_y_pos += mini_boss_direction * 0.5  # Move up or down
            if mini_boss_mouse_mdas_y_pos <= 170:  # Move up limit
                mini_boss_direction = 0.5
            elif mini_boss_mouse_mdas_y_pos >= 190:  # Move down limit
                mini_boss_direction = -0.5
            # Check the elapsed time to control mouse visibility
            if not mouse_mdas_visible and (pygame.time.get_ticks() - mouse_mdas_hide_start_time >= mouse_mdas_hide_duration):
                mouse_mdas_visible = True

            # Draw the mini boss mouse if it's visible
            if mouse_mdas_visible:
                screen.blit(mini_boss_mouse_mdas, (mini_boss_mouse_mdas_x_pos, mini_boss_mouse_mdas_y_pos))

            # Wizard Cat
            if activate_streak:
                screen.blit(wizard_bubble, (wizard_cat_x_pos, 250))
            else:
                screen.blit(wizard_cat,
                            (wizard_cat_x_pos, 370))  # Wizard Cat Position

            if streak_message_shown:
                streak_text = answer_font.render("You're on a streak!", True, white_color)
                screen.blit(streak_text, (275, 30))

            if wizard_moving:
                current_time = pygame.time.get_ticks()  # Get current time in milliseconds
                wizard_cat_x_pos += 10  # Adjust this value for speed

                # Check if the movement duration has passed
                if current_time - wizard_move_start_time >= wizard_move_duration:
                    wizard_moving = False  # Stop moving
                    wizard_cat_x_pos = 30  # Reset to starting position
                    streak_message_shown = False

            # Question and Answer
            question_surface = question_font.render(mdas_question, True, white_color)
            input_surface = title_font.render(user_input, True, white_color)

            # mdas Question
            if mdas_question is None:
                mdas_question, mdas_answer = generate_mdas_question()
            if not mdas_question_visible and (pygame.time.get_ticks() - mdas_hide_start_time >= mdas_hide_duration):
                mdas_question_visible = True
            if mdas_question_visible:
                # Draw question and user input
                screen.blit(question_surface, ((screen_width - question_surface.get_width()) // 2 + 10, 75))
                screen.blit(input_surface, (answer_input_box.x + (answer_input_box.width - input_surface.get_width()) // 2, answer_input_box.y + (answer_input_box.height - input_surface.get_height()) //2))

            # Input Answer Box
            if not input_box_visible and (pygame.time.get_ticks() - input_box_hide_start_time >= input_box_hide_duration):
                input_box_visible = True
            # Render question and user input
            if input_box_visible:
                pygame.draw.rect(screen, (255, 255, 255, 20), answer_input_box, 2)

            # Draw the correct answer message if applicable
            if show_correct_answer:
                screen.blit(correct_answer_message, ((screen_width - correct_answer_message.get_width()) // 2, 150))

        elif current_state == SHRINKING_MICE:
            # Shrink each image size only if above minimum size
            if addition_current_width > minimum_size and addition_current_height > minimum_size:
                addition_current_width -= shrink_amount
                addition_current_height -= shrink_amount

            if subtraction_current_width > minimum_size and subtraction_current_height > minimum_size:
                subtraction_current_width -= shrink_amount
                subtraction_current_height -= shrink_amount

            if multiplication_current_width > minimum_size and multiplication_current_height > minimum_size:
                multiplication_current_width -= shrink_amount
                multiplication_current_height -= shrink_amount

            if division_current_width > minimum_size and division_current_height > minimum_size:
                division_current_width -= shrink_amount
                division_current_height -= shrink_amount

            if mdas_current_width > minimum_size and mdas_current_height > minimum_size:
                mdas_current_width -= shrink_amount
                mdas_current_height -= shrink_amount

            # Scale the images
            addition_scaled = pygame.transform.scale(mini_boss_mouse_addition,(addition_current_width, addition_current_height))
            subtraction_scaled = pygame.transform.scale(mini_boss_mouse_subtraction,(subtraction_current_width, subtraction_current_height))
            multiplication_scaled = pygame.transform.scale(mini_boss_mouse_multiplication,(multiplication_current_width, multiplication_current_height))
            division_scaled = pygame.transform.scale(mini_boss_mouse_division,(division_current_width, division_current_height))
            mdas_scaled = pygame.transform.scale(mini_boss_mouse_mdas,(mdas_current_width, mdas_current_height))

            screen.blit(addition_bg, (0, 0))
            screen.blit(addition_scaled, (50, 300))
            screen.blit(subtraction_scaled, (235, 300))
            screen.blit(multiplication_scaled, (425, 300))
            screen.blit(division_scaled, (600, 300))
            screen.blit(mdas_scaled, (800, 300))

            text_surface = title_font.render("YOU ARE AN", True, black_color)
            text_surface2 = title_font.render("ACCOMPLISHED WIZARD!", True,black_color)
            screen.blit(text_surface,((screen_width - text_surface.get_width()) // 2, 75))
            screen.blit(text_surface2,((screen_width - text_surface2.get_width()) // 2, 125))

            text_surface3 = smaller_font.render("Press [C] to Continue", True,white_color)
            screen.blit(text_surface3, ((725, 460)))

        ### GAME OVER ###
        elif current_state == GAME_OVER:
            game_over_screen()
            continue

        ### LEVEL_COMPLETED ###
        # Check if the user has answered enough questions correctly
        elif current_state == LEVEL_1_COMPLETED:
            print('Level 1 Completed')
            level_1_completed_screen()
            current_state = LOAD_LEVEL_2

        elif current_state == LEVEL_2_COMPLETED:
            print('Level 2 Completed')
            level_2_completed_screen()
            current_state = LOAD_LEVEL_3

        elif current_state == LEVEL_3_COMPLETED:
            print('Level 3 Completed')
            level_3_completed_screen()
            current_state = LOAD_LEVEL_4

        elif current_state == LEVEL_4_COMPLETED:
            print('Level 4 Completed')
            level_4_completed_screen()
            current_state = LOAD_LEVEL_5

        elif current_state == LEVEL_5_COMPLETED:
            print('BOSS DEFEATED! YOU WIN!')
            level_5_completed_screen()
            current_state = SHRINKING_MICE

        elif current_state == GAME_ACCOMPLISHED:
            game_accomplished_screen()
            continue

    input_box.entered = False  # Reset the entered flag

    menu_button.update()  

    pygame.display.update()

# Save usernames back to file before quitting
save_usernames('used_usernames.pkl', used_usernames)

pygame.quit()
```
